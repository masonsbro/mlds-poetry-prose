```python
import matplotlib.pyplot as plt
import pandas as pd
from pandas.api.types import is_numeric_dtype
import numpy as np
import random
import collections
```

```python
df = pd.read_csv('training_dataset.csv')
```

```python
def remove_src(line):
    if not line:
        return ''
    try:
        return line[line.index('>') + 2:]
    except:
        global errors
        errors += 1
        return line
```

```python
def clean_text(text):
    lines = text.split('\n')
    lines = [remove_src(line) for line in lines]
    return ''.join([line + '\n' for line in lines if line])
```

```python
def mean_line_length(text):
    lines = text.split('\n')
    return np.mean(list(map(len, lines)))

def std_line_length(text):
    lines = text.split('\n')
    return np.std(list(map(len, lines)))

def filter_word(word):
    return ''.join([c for c in word if c in 'abcdefghijklmnopqrstuvwxyz'])

def get_words(text):
    return list(map(filter_word, text.lower().split()))

def common_words(text):
    return [k for k, v in collections.Counter(get_words(text)).most_common(25)]
```

```python
def all_common_words(df):
    cur = set()
    for i, row in df.iterrows():
        cur |= set([filter_word(word) for word in common_words(row['text'])])
    if '' in cur:
        cur.remove('')
    return list(cur)
```

```python
def preprocess(df):
    df = df.copy()
    global errors
    errors = 0
    df['clean'] = df['text'].apply(clean_text)
    print('Cleaning errors: {}'.format(errors))
    df['mean_line_length'] = df['clean'].apply(mean_line_length)
    df['std_line_length'] = df['clean'].apply(std_line_length)
    df['std_line_length_rat'] = df['std_line_length'] / df['mean_line_length']
    return df
```

```python
df = preprocess(df)
```

```python
important_words = all_common_words(df)
```

```python
len(important_words)
```

```python
def count_important_words(row):
    words = get_words(row['text'])
    d = len(words)
    c = collections.Counter(words)
    return [1.0 * c[w] / d for w in important_words]
```

```python
tmp = {}
for i, row in df.iterrows():
    tmp[row['ID']] = count_important_words(row)

important_word_cols = ['word_' + w for w in important_words]
df = df.assign(**{w: df['ID'].apply(lambda x: tmp[x][i]) for i, w in enumerate(important_word_cols)})
del tmp
```

```python
df = df.sample(frac=1)

tr = df[:350]
va = df[350:]
```

```python
def upsample_minority(tr):
    max_class_cnt = tr['class'].value_counts().max()
    for cl in ['prose', 'poetry']:
        cur_class_cnt = len(tr[tr['class'] == cl])
        if cur_class_cnt < max_class_cnt:
            tr = tr.append(tr[tr['class'] == cl].sample(n=(max_class_cnt - cur_class_cnt)))
    return tr

tr = upsample_minority(tr)
```

```python
forbidden = ['ID', 'mean_line_length', 'std_line_length', 'std_line_length_rat']
all_attrs = [attr for attr in list(tr) if is_numeric_dtype(tr[attr])]
attrs = [attr for attr in all_attrs if attr not in forbidden and not attr.startswith('word_')]
classes = ['prose', 'poetry']
fig = plt.figure(figsize=(8 * len(classes), 4 * len(attrs)))
for r, attr in enumerate(attrs):
    for c, cl in enumerate(classes):
        ax = fig.add_subplot(len(attrs), len(classes), r * len(classes) + c + 1)
        ax.hist(tr[tr['class'] == cl][attr], bins=np.linspace(tr[attr].min(), tr[attr].max(), 20))
        ax.set_title('{} {}'.format(cl, attr))
    if r >= 3: break
fig.savefig('pdfs.png', bbox_inches='tight', pad_inches=0)
```

```python
indiv_acc = {}
assert len(classes) == 2
for r, attr in enumerate(attrs):
    pts = [(row[attr], random.random(), row['class']) for i, row in tr.iterrows()]
    pts.sort()
    class_cnts = {}
    for cl in classes:
        class_cnts[cl] = tr[tr['class'] == cl].count()[0]
    pre = {cl: 0 for cl in classes}
    best = 0.0
    for i in range(len(pts)):
        # pre[class1] = success type 1
        # 1 - pre[class2] = success type 2
        # success = 1/2 * (1 - pre[class2] + pre[class1])
        pre[pts[i][2]] += 1
        for which in range(2):
            cur = 0.5 * (1 - 1.0 * pre[classes[which]] / class_cnts[classes[which]] +
                         1.0 * pre[classes[not which]] / class_cnts[classes[not which]])
            best = max(best, abs(cur))
    indiv_acc[attr] = best

items = [(v, k) for k, v in indiv_acc.items()]
items.sort(reverse=True)

for v, k in items:
    print(k, '{:0.04f}'.format(v))
```

```python
from sklearn import tree
from sklearn.model_selection import cross_val_score
import graphviz

def build_clf(pred_attrs):
    X = tr[pred_attrs]
    Y = tr['class']
    clf = tree.DecisionTreeClassifier()
    clf = clf.fit(X, Y)
    return clf

def evaluate_attrs(pred_attrs, clf):
    return clf.score(va[pred_attrs], va['class'])

def graph_attrs(pred_attrs, clf, out_file):
    tree.export_graphviz(clf, out_file=out_file, feature_names=pred_attrs)
    return tree.export_graphviz(clf, out_file=None, feature_names=pred_attrs)
```

```python
pred_attrs = attrs[:]
print(pred_attrs)
clf = build_clf(pred_attrs)
print('score: {:.4f}'.format(evaluate_attrs(pred_attrs, clf)))
graphviz.Source(graph_attrs(pred_attrs, clf, 'hand_engineered_tree.dot'))
```

```python
pred_attrs = ['word_eius', 'word_in', 'word_eorum', 'word_ex']
print(pred_attrs)
clf = build_clf(pred_attrs)
print('score: {:.4f}'.format(evaluate_attrs(pred_attrs, clf)))
graphviz.Source(graph_attrs(pred_attrs, clf, 'brute_force_tree.dot'))
```
